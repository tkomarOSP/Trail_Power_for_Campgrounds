```python
import capellambse
import pandas as pd

from IPython.core.display import HTML

resources = {
    "erod_2024": "erod_2024/eRod_Autonomous_Vehicle_2024",
}
path_to_model = "../eRod_Autonomous_Vehicle_2024.aird"
model = capellambse.MelodyModel(path_to_model, resources=resources, jupyter_untrusted=False)
```

```python
OP = model.search("OperationalProcess").by_name("Remote Park")

#print(OP.involved_functions)

table_data = {
        'Owning Entity': [],
        'Function': []
}
for function in OP.involved_functions:
    #help(function)
    #print(function.owner.name,function.name)
 
    table_data['Owning Entity'].extend([function.owner.name])
    table_data['Function'].extend([function.name])
    
df = pd.DataFrame(table_data)
print (df)








```

```python
#PA = model.search("PhysicalArchitecture")
#print(PA.all_functional_chains)
for fc in model.pa.all_functional_chains:
    #print(fc)
    table_data = {
        'Owning Entity': [],
        'Function': [],
        'Inputs': [],
        'Outputs': []
        }
    print("Functional Chain :", fc.name)
    for function in fc.involved_functions:
        #print(function)
        
        #print(function.owner,function.name)
        if function.owner != None :
            table_data['Owning Entity'].extend([function.owner.name])
            table_data['Function'].extend([function.name])
            #I really only want the funtional exchange that is involved
            table_data['Inputs'].extend([function.inputs])
            table_data['Outputs'].extend([function.outputs])
        else: 
            table_data['Owning Entity'].extend(["None"])
            table_data['Function'].extend([function.name])
            table_data['Inputs'].extend([function.inputs])
            table_data['Outputs'].extend([function.outputs])
        
    df = pd.DataFrame(table_data)

    display(df)





```
